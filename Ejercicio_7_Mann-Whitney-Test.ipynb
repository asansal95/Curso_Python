#Ejercicio 7: Nonparametric Mann-Whitney U Test

```python
def import_csv(file):
    import pandas as pd
    return pd.read_csv(file)
```

```python
path = "../Datos/Terminos_lagoon_TA_DIC_2023_RawData.csv"

data_frame = import_csv(path)
```

```python
from scipy.stats import mannwhitneyu

def mann_whitney_test(data1, data2, alpha=0.05):
    """
    Performs the Mann-Whitney U test and prints U statistic, p-value, and result.
    """
    result = mannwhitneyu(data1, data2, alternative='two-sided')
    
    print(f"U statistic = {result.statistic}")
    print(f"p-value = {result.pvalue}")

    if result.pvalue < alpha:
        print("The difference is statistically significant.")
    else:
        print("The difference is not statistically significant.")
```

### DIC 

##### Boxplot DIC por temporada

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.boxplot(x='season', y='dic_micromol_kg', data=data_frame)
#sns.swarmplot(x='season', y='sal_psu', data=co2_data, color=".25")

plt.xlabel('Season')
plt.ylabel('DIC (μmol/kg)')
plt.title('Box Plot of DIC by Season')

plt.show()
```

##### Violin Plot DIC por temporada

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Create a box plot
sns.violinplot(x='season', y='dic_micromol_kg', data=data_frame)

plt.title('DIC Distribution by Season')
plt.xlabel('Season')
plt.ylabel('DIC (μmol/kg)')
plt.show()
```

##### Mann-Whitney U Test para DIC

```python
dic_dry = data_frame.loc[data_frame["season"] == "Dry", "dic_micromol_kg"]
dic_rainy = data_frame.loc[data_frame["season"] == "Rainy", "dic_micromol_kg"]

mann_whitney_test(dic_dry, dic_rainy)
```
